```python
import numpy as np
matrix = np.array([[4,2,1],
                   [3,6,5],
                   [9,7,8]])
new_matrix = matrix *2
#calculating the transpose
print("The tranpose of the matrix is:")
matrix_transpose = new_matrix.T
print(matrix_transpose)
#calculating the determinant of the original matrix
print("The determinant of the original matrix is:")
det_matrix = np.linalg.det(matrix)
print(det_matrix)

```

```python
#RESERVOIR PRESSURE SIMULATION
import numpy as np
min_pressure = 2000
max_pressure = 5000
reservoir_grid = np.random.uniform(min_pressure,max_pressure,(10,10))
print("The reservoir grid is\n",reservoir_grid)


#flatten the array to 1D first to be able to sum it all
flattened_array = reservoir_grid.flatten()
cumulative_sum = np.cumsum(flattened_array)

#the cum sum is the last number at the end of the sum because it is cumulative and everything sums up to the end
#calculating the mean pressure
total_sum = cumulative_sum[-1]
reservoir_grid_mean = total_sum/100
print(f"The reservoir mean pressure is {reservoir_grid_mean:.3f} psi") 
#or use np.mean()

#calculating the minimum pressure
reservoir_grid_min = min(flattened_array)
print(f"The minimum pressure in the reservoir is {reservoir_grid_min:.3f} psi")
#or use np.min()

#calculating the maximum pressure
reservoir_grid_max = max(flattened_array)
print(f"The maximum pressure in the reservoir is {reservoir_grid_max:.3f} psi")
#or use np.max()

#find the index of the maximum value in the flattened array
max_index_flat = np.argmax(reservoir_grid)

#convert the flattened index to a 2d index(row,colum)
max_index_2d = np.unravel_index(max_index_flat,reservoir_grid.shape)
print("The location of the maximum pressure point is:",max_index_2d)

#replacing all pressures greater than 4500 using index method
reservoir_grid[reservoir_grid > 4500] = 4500
print(reservoir_grid)
        
        








```

```python
#FLUID FLOW IN PIPELINE GRID
import numpy as np
velocity_range_a = 2
velocity_range_b = 15

pipeline_grid = np.random.uniform(velocity_range_a,velocity_range_b,(6,6))

#calculating average velocity in the pipeline
pipeline_avg_vel = np.mean(pipeline_grid)
print(f"The average velocity in the pipeline is {pipeline_avg_vel:.3f} ft/s")

#counting how many grid cells have velocity above 12ft/s
print("The number of grid cells greater than 12 ft/s is:",len(pipeline_grid[pipeline_grid>12]))

new_grid = pipeline_grid.flatten()
for grid in new_grid:
    if grid < 5:
        print("Low")
    elif grid >=5 and grid <=12:
        print("Normal")
    elif grid >12:
        print("High")
    else:
        ("Not within range")

```

```python
#POROSITY CALCULATION FOR CORE SAMPLES
import numpy as np
#BULK VOLUMES
upper_bv = 300
lower_bv = 150

#PORE VOLUMES
upper_pv = 120
lower_pv = 20

#SIMULATING THE DATA FOR 20 CORE SAMPLES
bulk_volume = np.random.uniform(upper_bv,lower_bv,(20))
pore_volume = np.random.uniform(upper_pv,lower_pv,(20))

#POROSITY FOR EACH SAMPLE
porosity = pore_volume/bulk_volume *100
print("The porosity of each sample is:", porosity)

#CALCULATING THE AVERAGE POROSITY
avg_porosity = np.mean(porosity)
print(f"The average porosity is {avg_porosity:.3f} %")

#NUMBER OF SAMPLES WITH POROSITY LESS THAN 20%
print("Number of samples with porosity less than 20 is:",len(porosity[porosity<20]))

import seaborn as sns
import matplotlib as plt
from numpy import random

sns.displot(porosity,kind="hist",bins=10,color='skyblue')
plt.title("Histogram of Porosity Values")
plt.xlabel("Porosity (%)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()


```

```python
#SIMULATED RESERVOIR PRESSURE DROP OVER TIME
import math
import numpy as np
import random

k = 0.1
t = 10

pressure_decline_grid = np.random.uniform(4000,5000,(10,10))
#pressure exponential after 10 years
pressure_expo = pressure_decline_grid*(np.exp(-k*t))

import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(pressure_expo,kind="kde",color='skyblue')
plt.title("Exponential decay")
plt.xlabel("pressure_expo")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

#average pressure across the grid
avg_expo_pressure = np.mean(pressure_expo)
print(f"Average final pressure across the grid is: {avg_expo_pressure:.3f} psi")

#how many grid point dropped below 1500 psi by year 10
print("Number of grids below 1500 psi is:",len(pressure_expo[pressure_expo<1500]))


```

```python
#grid transformation
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pressure_matrix = np.random.randint(1000,3000,(6,6))
print("-----ORIGNAL MATRIX-----")
print(pressure_matrix)

#REPLACING ALL VALUES
pressure_matrix[pressure_matrix < 1500] = 0
pressure_matrix[(pressure_matrix >=1500) & (pressure_matrix <=2500)] = 1
pressure_matrix[pressure_matrix > 2500] = 2
print("-----TRANSFORMED MATRIX-----")
print(pressure_matrix)

print("--------------------------------")
#COUNTING ALL VALUES THAT FALL INTO EACH CATERGORY
print("Number of grid cells with zero is:",np.sum(pressure_matrix ==0))
print("Number of grid cells with one is:",np.sum(pressure_matrix ==1))
print("Number of grid cells with two is:",np.sum(pressure_matrix ==2))

print("-----SEA-BORN HEATMAP-----")
sns.displot(pressure_matrix,kind="kde",color='skyblue')
plt.xlabel("pressure_matrix")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```
