```python
from scipy.integrate import quad
import numpy as np

#compute the intergral of sin(x), with lower limit of 0 and upper limit of pi
result = quad(lambda x:np.sin(x),0,np.pi)
print(f"The result is: {result}")
#print(f"The error is: {error}")
    
    
```

```python
#SPECIAL FUNCTIONS
import scipy.special
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,500)

sigmoid = lambda x:1/(1 + np.exp(-x))
y = sigmoid(x)

#VISUALISATION
plt.plot(x, y, label="Sigmoid Curve")
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("σ(x)")
plt.grid(True)
plt.legend()
plt.show()




```

```python
#CLASSIC GAUSSAIAN DISTRIBUTION CURVE
import numpy as np
x = np.linspace(-5,5,500)
#defining the gaussian function
meuw = 0
sigma = 1
gaussian_function = lambda x: 1/(sigma*np.sqrt(2*np.pi)) * (np.exp(-(x-meuw)**2/(2*sigma**2)))
numpy_gaussian = gaussian_function(x)

import matplotlib.pyplot as plt
plt.plot(x,numpy_gaussian,label = "GAUSSAIAN DISTRIBUTION CURVE")
plt.title("CLASSIC GAUSSAIAN DISTRIBUTION")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
#COMPARING GUASSIAN NUMPY GRAPH AND SCIPY GUASSIAN GRAPH
# plot 1: numpy guassian curve
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].plot(x,numpy_gaussian,label = "GAUSSAIAN DISTRIBUTION CURVE")
axs[0].set_title("CLASSIC GAUSSAIAN DISTRIBUTION")
axs[0].set_xlabel("x")
axs[0].set_ylabel("y")
axs[0].legend()
axs[0].grid()


# plot 2: scipy gaussain curve
from scipy.stats import norm
gaussian_scipy = norm.pdf(x,loc=meuw,scale=sigma)
axs[1].plot(x,gaussian_scipy,label ="SCIPY GAUSSAIAN DISTRIBUTION CURVE")
axs[1].set_title("CLASSIC GAUSSAIAN DISTRIBUTION")
axs[1].set_xlabel("x")
axs[1].set_ylabel("y")
axs[1].legend()
axs[1].grid()

plt.tight_layout()
plt.show()



```
