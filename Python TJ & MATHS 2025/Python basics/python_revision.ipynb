```python
#ENCRYPTION PROGRAM IN PYTHON
#import random
import string
chars = " " +string.punctuation + string.digits + string.ascii_letters
chars = list(chars)
key = chars.copy() #COPY THE OUTPUT FOR CHARS TO KEY

random.shuffle(key)

print(f"chars: {chars}")
print(f"Key:    {key}")

#ENCRYPTION
real_word = input("Enter your message to encrypt")
cipher_text = ""
for letter in real_word:
    index = chars.index(letter)
    cipher_text += key[index]

print(f"original_message: {real_word}")
print(f"cipher_text: {cipher_text}")
```
