```python
#FILTERING ARRAYS
import numpy as np
arr = np.array([40,41,42,43,44,45])
new_arr = []
for number in arr:
    if number >41:
        new_arr.append(True)
    else:
        new_arr.append(False)
print(new_arr)
news_arr = arr[new_arr]
print(news_arr)
        
```

```python
import numpy as np

arr = np.array([41, 42, 43, 44])

filter_arr = arr > 42

newarr = arr[filter_arr]

print(filter_arr)
print(newarr)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#PLOTTING A HISTOGRAM
sns.displot([0, 1, 2, 3, 4, 5], kind="hist")

plt.show()

#PLOTTING WITHOUT USING A HISTOGRAM
sns.displot([0, 1, 2, 3, 4, 5], kind="kde")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])
#PLOTTING WITH THE LINE
plt.plot(xpoints, ypoints)
plt.show()
#PLOTTING WITHOUT THE LINE
plt.plot(xpoints, ypoints, "o") #USING THE "o" PARAMETER
plt.show()
```
